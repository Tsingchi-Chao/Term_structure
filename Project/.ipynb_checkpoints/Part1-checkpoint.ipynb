# Part 1

From the FRN in Table1, it's obvious that $y_{1}=0.206\%$.

$y_x$ means the yield to maturity of the $0.5x$-year zero-coupon bond.
Here, $y_{1}$ means the yield to maturity for the $0.5$-year zero-coupon bond.

According to Chapter 05-part 1,

$$ R_{m,n}(t)=\frac{P(t,T_m)-P(t,T_n)}{\sum_{j=m}^{n-1} \Delta T_j P(t,T_{j+1})}$$
Here, $R_{m,n}(t)$ is the swap rate for the interest rate swap.

For the interest swap in Table1, $T_m=0,t=0,\Delta T_j=0.5$, so

$$R_{m,n}(t)=\frac{1-P(0,T_n)}{\Delta_{j=m}^{n-1}0.5*P(0,T_{j+1})}$$

```python
def P_Tn(yn:float,n:int)->float:
    """
    To calculate the present value.
    parameter
    param yn: The YTM for the zero-coupon bond.
    param n: The number of the 0.5-year.
    """
    return 1/((1+yn*0.5)**n)
```

```python
Term=[2,3,4,5,7,10,20,30]
swap_rate=[0.00202,0.00288,0.00434,0.00597,0.00919,0.01263,0.017,0.01788]
Rmn = dict(zip(Term, swap_rate))
ytm=OrderedDict()
ytm[1]=0.00206
```

```python
def f(yn:float,n:int,Rmn:dict,ytm:dict)->float:
    """
    The function we want to solve.
    parameters
    param yn: The ytm of the 0.5*n-year zero-coupon bond.
    param n: The number of the 0.5-year.
    param Rmn:The dictionary for the interest swap.
    param ytm:The dictionary for the yield to maturity that have been solved.
    """
    numerator=1-P_Tn(yn,n)
    denominator=0
    for i in ytm.keys():
        denominator+=0.5*P_Tn(ytm[i],i)
    ym=ytm[list(ytm.keys())[-1]]
    m=list(ytm.keys())[-1]
    for i in range(list(ytm.keys())[-1]+1,n+1):
        y=ym+(yn-ym)*(i-m)/(n-m)
        denominator+=0.5*P_Tn(y,i)
    return numerator/denominator-Rmn[int(n/2)]
```

```python
def bisection(low_bound:float,up_bound:float,epsilon:float,n:int,Rmn:dict,ytm:dict)->float:
    """
    parameters
    param low_bound: The low bound you set.
    param up_bound: The up bound you set.
    param epsilon: The error.
    param n: The number of the 0.5-year.
    param Rmn:The dictionary for the interest swap.
    param ytm:The dictionary for the yield to maturity that have been solved.
    """
    for i in range(100):
        y=(low_bound+up_bound)/2
        if abs(f(y,n,Rmn,ytm))<epsilon:
            ym=ytm[list(ytm.keys())[-1]]
            m=list(ytm.keys())[-1]
            for j in range(list(ytm.keys())[-1]+1,n+1):
                ytm[j]=ym+(y-ym)*(j-m)/(n-m)
            return ytm
        else:
            if f(y,n,Rmn,ytm)*f(up_bound,n,Rmn,ytm)<0:
                low_bound=y
            elif f(y,n,Rmn,ytm)*f(low_bound,n,Rmn,ytm)<0:
                up_bound=y
```

```python
for term in Term:
    ytm=bisection(0.001,0.1,1e-10,2*term,Rmn,ytm)
data=pd.DataFrame()
for i in ytm.keys():
    data.loc[i/2,'YTM']=ytm[i]
    data.loc[i/2,'discount_rate']=P_Tn(ytm[i],i)
```

```python
def get_ytm(x:float,ytm:dict):
    """
    The function of yield to maturity.
    parameter
    param x:The number of 0.5-year.
    param ytm:The dictionary of ytm.
    """
    if x<1:
        return ytm[1]
    if x>=1 and x<=4:
        return ytm[1]+((ytm[4]-ytm[1])/(4-1))*(x-1)
    elif x>4 and x<=6:
        return ytm[4]+((ytm[6]-ytm[4])/(6-4))*(x-4)
    elif x>6 and x<=8:
        return ytm[6]+((ytm[8]-ytm[6])/(8-6))*(x-6)
    elif x>8 and x<=10:
        return ytm[8]+((ytm[10]-ytm[8])/(10-8))*(x-8)
    elif x>10 and x<=14:
        return ytm[10]+((ytm[14]-ytm[10])/(14-10))*(x-10)
    elif x>14 and x<=20:
        return ytm[14]+((ytm[20]-ytm[14])/(20-14))*(x-14)
    elif x>20 and x<=40:
        return ytm[20]+((ytm[40]-ytm[20])/(40-20))*(x-20)
    elif x>40 and x<=60:
        return ytm[40]+((ytm[60]-ytm[40])/(60-40))*(x-40)
    
def get_discount_rate(x:float,ytm:dict):
    """
    To calculate discount rate.
    parameter
    param x:The number of 0.5-year.
    """
    discount_rate=P_Tn(get_ytm(x,ytm),x)
    return discount_rate
```

## lattice tree

```python
K=0.12
sigma=0.035
delta_T=0.5
delta_t=delta_T/2
delta_r=sigma*math.sqrt(3*delta_t)
J=math.floor(1/(2*K*delta_t))+1
```

```python
def get_k(j:int):
    """
    This function is calculating the k after state j.
    parameter
    param j: The state.
    """
    if j==-J:
        return -J+1
    elif abs(j)<J:
        return j
    elif j==J:
        return J-1
    else:
        print('There is something wrong,please check it.')
    
def get_eta(j:int,n:int):
    """
    This function is to calculating eta of j,n.
    parameter
    param j:The state.
    param n:The step.
    """
    eta_j_n=((1-K*delta_t)*j-get_k(j))*delta_r
    return eta_j_n

def get_p(j:int,n:int,x:int):
    """
    This function is calculating probability by j,n,x and x may be k-1,k or k+1.
    parameter
    param j:The state.
    param n:The step.
    param x:It may be k-1,k or k+1.
    """
    k=get_k(j)
    if x==k+1:
        p=0.5*(((sigma**2)*delta_t+(get_eta(j,n))**2)/(delta_r**2)+get_eta(j,n)/delta_r)
        return p
    elif x==k-1:
        p=0.5*(((sigma**2)*delta_t+(get_eta(j,n))**2)/(delta_r**2)-get_eta(j,n)/delta_r)
        return p
    elif x==k:
        p=1-((sigma**2)*delta_t+(get_eta(j,n))**2)/(delta_r**2)
        return p
    else:
        p=0
        return p
```

```python
Q=OrderedDict()
Q[0]=OrderedDict()
Q[0][0]=1
r=OrderedDict()
r[0]=OrderedDict()
r[0][0]=0.00206
theta=OrderedDict()
```

```python
n=math.floor(30/delta_t)
for j in range(1,n):
    Q[j]=OrderedDict()
    r[j]=OrderedDict()     
    for m in range(-1*min(j,J),min(j,J)+1):
        Q[j][m]=0
        for i in range(-1*min(j-1,J),min(j-1,J)+1):
            Q[j][m]+=Q[j-1][i]*get_p(i,j-1,m)*math.exp(-r[j-1][i]*delta_t)    
    for i in range(-1*min(j,J),min(j,J)+1):
        r[j][i]=(1-K*delta_t)*r[j-1][0]+i*delta_r
    s=0
    for i in range(-1*min(j,J),min(j,J)+1):
        s+=Q[j][i]*math.exp(-1*r[j][i]*delta_t)
    theta[j-1]=(1/(K*(delta_t**2)))*math.log(s/get_discount_rate(((j+1)*delta_t/0.5),ytm))
    
    for i in range(-1*min(j,J),min(j,J)+1):
        r[j][i]=r[j][i]+K*theta[j-1]*delta_t
```

```python
df1=pd.DataFrame()
for i in range(n-1):
    df1.loc[i*delta_t,'theta']=theta[i]
```

```python
plt.plot(df1)
plt.title('Theta')
# plt.axis('Year')
```

```python
df2=pd.DataFrame()
for i in range(1,n-1):
    for j in range(-J,J+1):
        try:
            df2.loc[i*delta_t,j]=r[i][j]
        except:
            continue
```

```python
fig = plt.figure(figsize=[10,6])
ax1= fig.add_subplot(1, 1, 1)
for i in range(-J,J+1):
    if i==0:
        ax1.plot(df2[i],"*",c='b')
    else:
        ax1.plot(df2[i],"-")
ax1.set_xlabel('year')
ax1.set_ylabel('short rate')
ax1.set_title('The Tree')
```

## Pricing

```python
T0=15  
T0_step=math.floor(T0/delta_t)   #The time step of T0
T_N=20
T_N_step=math.floor(T_N/delta_t)
r_up_bound=0.2
r_low_bound=-0.2
c1=1
c2=2
c3=3
strike_price=10
```

```python
#To get the coupon.
for j in range(T_N_step,-1,-1):
    coupon[j]=OrderedDict()
    if (T_N_step-j)%(delta_T/delta_t)==0:  #Because delta_T and delta_t is not equal.
        for i in range(-min(j,J),min(j,J)+1):
            if r[j][i]>=r_up_bound:
                coupon[j][i]=c1*delta_T
            elif r[j][i]>=r_low_bound:
                coupon[j][i]=c2*delta_T
            elif r[j][i]>=0:
                coupon[j][i]=c3*delta_T
            else:
                coupon[j][i]=0
    else:
        for i in range(-min(j,J),min(j,J)+1):
            coupon[j][i]=0

        
for j in range(-min(T_N_step,J),min(T_N_step,J)+1):   #State.
    Btc[T_N_step][j]=1

for j in range(T_N_step-1,-1,-1):   # Step.
    Btc[j]=OrderedDict()
    for i in range(-min(j,J),min(j,J)+1):  #State.
        Btc[j][i]=0
        for k in range(-min(j+1,J),min(j+1,J)+1):
            Btc[j][i]+=(Btc[j+1][k]+coupon[j+1][k])*get_p(i,j,k)*math.exp(-r[j][i]*delta_t)
```

### European option pricing 

```python
V=OrderedDict() #V is price of the option. 
V[T0_step]=OrderedDict()
for i in range(-min(T0_step,J),min(T0_step,J)+1):
    V[T0_step][i]=max(Btc[T0_step][i]-strike_price,0)
```

```python
for j in range(T0_step-1,-1,-1):
    V[j]=OrderedDict()
    for i in range(-min(j,J),min(j,J)+1):
        V[j][i]=0
        for k in range(-min(j+1,J),min(j+1,J)+1):
            V[j][i]+=V[j+1][k]*get_p(i,j,k)*math.exp(-r[j][i]*delta_t)
```

```python
print('The option price at step 0 is:',V[0][0])
```

### American option pricing

```python
for j in range(T0_step-1,-1,-1):
    V[j]=OrderedDict()
    for i in range(-min(j,J),min(j,J)+1):
        V[j][i]=0
        for k in range(-min(j+1,J),min(j+1,J)+1):
            V[j][i]+=V[j+1][k]*get_p(i,j,k)*math.exp(-r[j][i]*delta_t)
        V[j][i]=max(V[j][i],Btc[j][i]-strike_price)
```

```python
print('The option price at step 0 is:',V[0][0])
```
